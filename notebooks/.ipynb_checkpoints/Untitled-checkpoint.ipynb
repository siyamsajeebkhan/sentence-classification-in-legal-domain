```python
file = open('../project_requirements.txt', 'r').readlines()
```

```python
file
```

```python
req_file_name = '../code/requirements_extended.txt'
req_file = open(req_file_name, 'w')
for line in file:
    package, version = line.split(" ")[0], line.split(" ")[-1]
    package_with_version = f"{package}=={version}"
    print(package_with_version)
    req_file.write(package_with_version)
req_file.close()    
```
